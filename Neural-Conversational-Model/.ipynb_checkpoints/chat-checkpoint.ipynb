```python
def plot_attention_map(model, 
                       batch, 
                       sample_idx,
                       min_attention_threshold=0.1):
    model.eval()
    with torch.no_grad():
    
        text_vecs = batch['text_vecs'].to(device)
        target_vecs = batch['target_vecs'].to(device)   
        target_seq_len = target_vecs.size(1)

        encoded = model.encoder(text_vecs, 
                                batch['text_lens'], 
                                use_packed=batch['use_packed'])

        decoder_output, preds, attn_w_log = model.decoder.decode_forced(
            target_vecs, encoded)

        attention_weights = []

        for i in range(target_seq_len):
            attention_weights.append(attn_w_log[i][sample_idx])

        attention_weights = torch.stack(attention_weights)

        sample_input_vec = text_vecs[sample_idx]
        sample_target_vec = target_vecs[sample_idx]
        sample_input_vec_mask = sample_input_vec.ne(0)
        sample_target_vec_mask = sample_target_vec.ne(0)
        sample_input_vec_last_idx = sample_input_vec_mask.nonzero(
            ).squeeze().max().item()
        sample_target_vec_last_idx = sample_target_vec_mask.nonzero(
            ).squeeze().max().item()
        sample_target_vec_shift = torch.cat([torch.Tensor([1]).long().to(device), 
                                       sample_target_vec[:sample_target_vec_last_idx]]) 

        attention_threshold_mask = attention_weights > min_attention_threshold
        attention_threshold_mask = attention_threshold_mask.any(dim=0)

        attention_threshold_mask_idxes = attention_threshold_mask.nonzero().squeeze()

        first_input_idx = attention_threshold_mask_idxes.min().item()
        last_idx = sample_input_vec_last_idx
        window_size = 1
        select_attention_weights = attention_weights[sample_target_vec_mask, 
                                                     first_input_idx - window_size:last_idx + 1]

        select_input_vec = sample_input_vec[first_input_idx - window_size:last_idx + 1]
        select_target_vec = sample_target_vec[sample_target_vec_mask]
        
        input_words = get_words_from_ids(select_input_vec)
        target_words = get_words_from_ids(sample_target_vec_shift)
        
        fig, ax = plt.subplots(figsize=(14, 7))
        
        sns.heatmap(select_attention_weights.to('cpu').numpy(), 
                    xticklabels=input_words,
                    yticklabels=target_words,
                    ax=ax)

    return
```

```python
def get_words_from_ids(id_tensor):
    id_array = id_tensor.to('cpu').numpy()
    
    words = []
    
    for word_id in id_array:
        word = chat_dict.ind2word[word_id]
    
        if word == '\n':
            words.append('\\n')
        else:
            words.append(word)
    return words
```

```python
def get_truncated_ids(token_ids):
    first_null_idx = token_ids.eq(0).nonzero().squeeze().min()
    truncated_token_ids = token_ids[:first_null_idx]
    return truncated_token_ids

def print_input_target_for_sample_index(batch, sample_idx):
    text_vec = batch['text_vecs'][sample_idx]
    target_vec = batch['target_vecs'][sample_idx]
    
    print('Input text: ')
    print(chat_dict.pred2text(get_truncated_ids(text_vec)))
    print('\n')
    
    print('Target text: ')
    print(chat_dict.pred2text(get_truncated_ids(target_vec)))
    print('\n')
```

```python
### add transformer encoder as optional encoder in seq2seq model.

# code below can help you to start it, but feel free to start from scratch

class EncoderTransformer(nn.Module):
    def __init__(self,
                 option,
                 shared_lt=None):
        super().__init__()
        if shared_lt is None:
            self.embed = nn.Embedding(option['vocab_size'], 
                                      option['embedding_dim'], 
                                      padding_idx=PADDING_IDX)
        else:
            self.embed = shared_lt
        self.position_embed = nn.Embedding(option['max_input_seq_len'], 
                                           option['embedding_dim'])
        
        encoder_layer = nn.TransformerEncoderLayer(option['embedding_dim'],
                                                   nhead=option['n_multiheads'],
                                                   dim_feedforward=option['transformer_ff_dim'],
                                                   dropout=option['transformer_dropout'])
        self.transformer = nn.TransformerEncoder(encoder_layer, 
                                                 num_layers=option['n_transformer_layers'])
        self.linear = nn.Linear(option['embedding_dim'], option['hidden_size'])
        

    def forward(self, text_vec):
        pos_seq = torch.arange(text_vec.size(1)).unsqueeze(0).to(device)
        pos_embedded = self.position_embed(pos_seq)
        embedded = self.embed(text_vec)
        
        embedded = pos_embedded + embedded  # apply pos embedding
        # embedded = (batch_size * length * embed_dim)
        # since transformer expects (length * batch_size * embed_dim) we transpose
        embedded_t = embedded.permute(1, 0, 2)
        
        # attention_mask should be True for padding indexes per documentation for
        # Transformer.forward
        #    "[src/tgt/memory]_key_padding_mask should be a ByteTensor where True values are positions
        #    that should be masked with float('-inf') and False values will be unchanged."
        attention_mask = text_vec.eq(PADDING_IDX)
        attention_mask_neg = ~attention_mask
        
        # output is also of shape (length * batch_size * embedding_dim)
        transformer_out = self.transformer(embedded_t, src_key_padding_mask=attention_mask)
        # Lets transpose again to be consistent with what DecoderRNN expects as input
        transformer_out_t = transformer_out.permute(1, 0, 2)
        
        result = F.relu(self.linear(transformer_out_t))
        
        return result, None, attention_mask_neg
```

```python
class MySeq2Seq(nn.Module):
    """
    Generic seq2seq model with attention mechanism.
    """
    def __init__(self, option):

        super().__init__()
        self.decoder = DecoderRNN(vocab_size=option['vocab_size'],
                                  embed_size=option['embedding_dim'],
                                  hidden_size=option['hidden_size'],
                                  num_layers=option['decoder_num_layers'],
                                  dropout=option['decoder_dropout'])
        self.encoder = EncoderTransformer(option,
                                          shared_lt=self.decoder.embedding)
        
    def train(self):
        self.encoder.train()
        self.decoder.train()
        
    def eval(self):
        self.encoder.eval()
        self.decoder.eval()
```

```python
def do_eval(model, valid_loader, criterion):

    val_loss = 0
    val_tokens = 0

    model.eval()

    with torch.no_grad():
        for i, batch in enumerate(valid_loader):

            text_vecs = batch['text_vecs'].to(device)
            target_vecs = batch['target_vecs'].to(device)

            encoded = model.encoder(text_vecs)

            decoder_output, preds, attn_w_log = model.decoder.decode_forced(target_vecs, 
                                                                            encoded,)

            scores = decoder_output.view(-1, decoder_output.size(-1))

            loss = criterion(scores, target_vecs.view(-1))

            num_tokens = target_vecs.ne(0).long().sum().item()

            val_tokens += num_tokens
            val_loss += loss.item()

        avg_val_loss = val_loss / val_tokens
    
    return avg_val_loss
```

```python
MAX_SENTENCE_LEN = 100
N_NUCLEUS_SAMPLES = 100
```

```python
def get_top_p_idxes_for_sample(all_probs, p_threshold):
    """Get top p idxes. Assumes all_probs is a 1-D tensor"""
    sorted_idxes = all_probs.argsort(descending=True)
    last_top_p_idx = (all_probs[sorted_idxes].cumsum(dim=-1) > p_threshold).nonzero().min().item()
    top_p_idxes = sorted_idxes[:last_top_p_idx + 1]
    return top_p_idxes

def nucleus_sample(all_probs, top_p_idxes):
    probs_for_top_p = all_probs[top_p_idxes]
    probs_for_top_p_normalized = probs_for_top_p / probs_for_top_p.sum()
    
    categorical_sample_idx = torch.multinomial(probs_for_top_p_normalized, 1)[0].item()
    sampled_top_p_idx = top_p_idxes[categorical_sample_idx]
    sampled_raw_prob = probs_for_top_p[categorical_sample_idx]

    return sampled_top_p_idx, sampled_raw_prob
```

```python
def nucleus_sample_decoding(model, 
                            batch, 
                            p_threshold,
                            n_samples=100,
                            verbose=True):

    text_vecs = batch['text_vecs']
    text_vec_lens = batch['text_lens']
    
    # Expect batch size of 1
    assert text_vecs.size(0) == 1
    assert isinstance(text_vec_lens, list)
    
    text_vecs = text_vecs.expand(n_samples, -1)
    text_vec_lens = text_vec_lens * n_samples
    
    if verbose:
        print(chat_dict.pred2text(text_vecs[0]))    
        
    model.eval()
    
    with torch.no_grad():
        
        # Initial conditions and variables
        starts = torch.Tensor([chat_dict.word2ind['__start__']]).long().expand(n_samples, 1).to(device)
        xs = starts
        
        finish_mask = torch.zeros(n_samples).byte().to(device)
        
        sequence_log_probs = torch.zeros(n_samples).float().to(device)
        all_preds = torch.zeros(n_samples, MAX_SENTENCE_LEN).long()
        
        encoded = model.encoder(text_vecs.to(device), text_vec_lens, use_packed=True)
        encoder_output, encoder_hidden, attention_mask = encoded
        decoder_hidden = encoder_hidden
        
        for i in range(MAX_SENTENCE_LEN):
            
            decoder_output, decoder_hidden, attn_w_log = model.decoder(xs, 
                                                                       decoder_hidden, 
                                                                       encoded)
            output_probs = torch.softmax(decoder_output, dim=-1).squeeze(dim=1)

            preds_for_step = []
            sequence_probs_for_step = []

            for j in range(output_probs.size(0)):
                
                if finish_mask[j].bool():
                    sampled_top_p_idx = torch.Tensor(
                        [chat_dict.word2ind['__end__']]).long().to(device)
                    sampled_prob = torch.Tensor([1]).float().to(device)
                    
                else:
                    output_probs_for_sample = output_probs[j]
                    top_p_idxes_for_sample = get_top_p_idxes_for_sample(output_probs_for_sample,
                                                                        p_threshold)

                    sampled_top_p_idx, sampled_prob = nucleus_sample(output_probs_for_sample,
                                                                     top_p_idxes_for_sample)
                
                preds_for_step.append(sampled_top_p_idx.view(-1))
                sequence_probs_for_step.append(sampled_prob.view(-1))

            preds_for_step = torch.cat(preds_for_step)
            xs = preds_for_step.view(-1, 1)
            all_preds[:, i] = preds_for_step
            
            sequence_probs_for_step = torch.cat(sequence_probs_for_step)
            sequence_log_probs += torch.log(sequence_probs_for_step)
            
            finish_mask += (preds_for_step == chat_dict.word2ind['__end__']).byte()
            if not (torch.any(~finish_mask.bool())):
                print('Terminating after {} steps'.format(i+1))
                break
            
    return all_preds, sequence_log_probs
```

```python
def truncate_sequence(sequence):
    """Truncate to show before end token"""
    mask = (sequence == chat_dict.word2ind['__end__'])
    
    if mask.sum().item() == 0:
        truncated_seq = sequence
    else:
        min_non_zero_idx = mask.nonzero().squeeze().min()
        truncated_seq = sequence[:min_non_zero_idx]
    return truncated_seq

def get_uniq_tokens(all_preds):
    uniq_tokens_list = all_preds.unique().tolist()
    # Take out null, start, end
    uniq_tokens = set(uniq_tokens_list) - set([0, 1, 2])
    return uniq_tokens
```

```python
def tokenize_text(text, lower=True):
    # Append '\n' at the beginning of each turn
    tokens = ['\n']
    
    for tok in nlp(text):
        token = tok.text
        if lower:
            token = token.lower()
            
        tokens.append(token)
    return tokens

def tokenize_text_as_ids(text):
    tokens = tokenize_text(text, lower=True)
    token_ids = chat_dict.t2v(tokens)
    token_ids = torch.tensor(token_ids, dtype=torch.long)
    return token_ids
```

```python
class Chatter:
    
    def __init__(self, 
                 persona_text, 
                 model,
                 nucleus_p=0.5):
        self.persona_text = persona_text
        self.model = model
        self.nucleus_p = nucleus_p
        
        self.persona_token_ids = self._parse_persona_text_to_ids(self.persona_text)
        self.turn_idx = 1
        self.accumulated_text_list = [self.persona_text]
        self.accumulated_id_list = [self.persona_token_ids]
        self.concatenated_text = self.persona_text
        self.concatenated_ids = self.persona_token_ids
        
    def _parse_persona_text_to_ids(self, persona_text):
        _inp_toked = RETOK.findall(persona_text)
        _inp_toked_id = chat_dict.t2v(_inp_toked)
        token_ids = torch.tensor(_inp_toked_id, dtype=torch.long)
        return token_ids
    
    def _update(self, text_ids):
        text_from_ids = chat_dict.pred2text(text_ids)
        self.accumulated_id_list.append(text_ids)
        self.accumulated_text_list.append(text_from_ids)
        
        self.concatenated_text += ' ' + text_from_ids
        self.concatenated_ids = torch.cat([self.concatenated_ids, text_ids])
    
    def input_text(self):
        user_input_text = input(
            'user input for turn {}: '.format(self.turn_idx))
        self.turn_idx += 1
        user_input_text_ids = tokenize_text_as_ids(user_input_text)
        self._update(user_input_text_ids)
    
    def output_decoding_chat(self, 
                             method='nucleus_sampling',
                             print_response=True,
                             show_history=True):
        if method == 'nucleus_sampling':
            sample_dict = {
                'text_vecs': self.concatenated_ids.view(1, -1),
                'text_lens': [self.concatenated_ids.size(0)]
            }

            sample_response, _ = nucleus_sample_decoding(self.model, 
                                                         sample_dict, 
                                                         self.nucleus_p,
                                                         n_samples=1,
                                                         verbose=False)
            sample_response = truncate_sequence(sample_response[0])
            
        else:
            raise ValueError('Unexpected decoding method {}'.format(method))
            
        if print_response:
            if show_history:
                print('====== input to decoding: entire history =======')
                print(self.concatenated_text)
                print('================================================\n')
            
            print('response: ', chat_dict.pred2text(sample_response))
            print('\n')
        
        # append '\n' at beginning of response
        sample_response = torch.cat([
            torch.tensor([chat_dict.word2ind['\n']], dtype=torch.long),
            sample_response])
        
        self._update(sample_response)
        return
```
